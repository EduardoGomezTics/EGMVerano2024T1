```python
# Datos de entrada
datos = ['Ramírez', 'García', 'González', 'Rodríguez', 'Pérez', 'Martínez', 'García', 'López', 'Gutiérrez', 'López',
    'Rodríguez', 'Martínez', 'Rodríguez', 'Hernández', 'García', 'García', 'González', 'Hernández', 'García', 'Gutiérrez',
    'Hernández', 'López', 'Pérez', 'González', 'Hernández', 'López', 'Martínez', 'García', 'Hernández', 'Sánchez',
    'Hernández', 'González', 'López', 'Rodríguez', 'Pérez', 'Sánchez', 'Hernández', 'González', 'García', 'Hernández',
    'Rodríguez', 'Martínez', 'Ramírez', 'Pérez', 'Sánchez', 'Hernández', 'González', 'García', 'Hernández', 'García',
    'Rodríguez', 'López', 'Sánchez', 'Martínez', 'López', 'García', 'Hernández', 'García', 'Hernández', 'López',
    'González', 'López', 'Martínez', 'López', 'García', 'Hernández', 'García', 'Hernández', 'Flores', 'Pérez',
    'González', 'López', 'Martínez', 'González', 'Martínez', 'Hernández', 'González', 'García', 'López', 'Sánchez',
    'Ramírez', 'Gutiérrez', 'Sánchez', 'Flores', 'García', 'Martínez', 'Flores', 'Pérez', 'Hernández', 'López',
    'Hernández', 'Martínez', 'García', 'Hernández', 'Flores', 'Pérez', 'González', 'Hernández', 'Sánchez', 'Martínez']

# Formatea los datos para evitar espacios y excepciones en cadena
datos_formated = fStr.format_str(datos)
print("Datos indexeados:", datos_formated)

# Ordena la lista de datos
datos_sorted = clas_str_sort.clases_str_sort(datos_formated)
print("Clases sort:", datos_sorted, "\n")

# Calcula la frecuencia absoluta de los datos
clases, fa_ab = abs.frec_abs(datos_sorted)

# Crea una copia de las clases originales sin ordenar
clases_orden = clases.copy()

# Obtiene las frecuencias absolutas ordenadas ascendente
fa_sorted = scFa.sort_clases_fa(clases_orden, orAsc.ordenar_asc(clases), fa_ab)

# Calcula la frecuencia relativa
frecuencia_rel = rel.frec_rel(fa_sorted)

# Calcula la frecuencia relativa acumulada
frecuencia_rel_acum = acum.frec_acum(frecuencia_rel)

# Imprime las clases, su frecuencia relativa y su frecuencia relativa acumulada ordenadas de forma ascendente
t.tabla(clases, fa_sorted, frecuencia_rel, frecuencia_rel_acum)

# Llama a las gráficas para ingresar sus parámetros

grafPie.plot_pie(fa_sorted, clases)  # Gráfico de pie
print("")



```

# Ejercicio 8

```python
# Datos de entrada
dt = [2.5, 4.5, 5.8, 2.2, 5.7,
5.3, 4.0, 0.9, 2.5, 2.3,
3.2, 4.9, 2.7, 3.7, 3.8,
4.9, 2.8, 3.6, 4.2, 3.1,
7.2, 1.9, 5.9, 3.3, 2.9,
1.1, 1.7, 2.6, 3.1, 4.4]
     
# Datos ordenados
dt = orAsc.ordenar_asc(dt)
print("Datos ordenados", dt)
# Obtenemos clases, lim inf, lim. sup, y marcas de clase
clases, lim_inf, lim_sup, mrks = cg.clases_grouped(dt)
print("\n")
print("Clases: ", clases)
print("Limites inferiores: ", lim_inf)
print("Limites superiores: ", lim_sup)
print("Marcas de clase: ", mrks)
print("\n")
# Frecuencia absoluta
fa = faGr.fa_grouped(lim_inf, lim_sup, dt)
print("Frecuencia absoluta", fa)
print("\n")
# Frecuencia relativa
fr = rel.frec_rel(fa)
print("Frecuencia relativa", fr)
print("\n")
# Frecuencia relativa acumulada
facum = acum.frec_acum(fr)
print("Frecuencia relativa acumulada", facum)
print("\n")
# Imprecion en tabla para todos los datos
tg.tabla_grouped(clases, lim_inf, lim_sup, mrks, fa, fr, facum)
# Llamada de las funciones para graficar
pHistograma.plot_grafHist(fr, mrks, clases)# Grafico Histograma
valores_ref_eje = list(range(1, len(mrks) + 1)) #  Marcas de clase en el eje x 
pOjiva.plot_grafOjiva(mrks, valores_ref_eje, facum) # Grafico de Ojiva
```
